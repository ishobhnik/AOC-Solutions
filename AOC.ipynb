```python
def is_safe_report(report):
    # Check the differences between consecutive levels
    for i in range(len(report) - 1):
        diff = abs(report[i] - report[i + 1])
        
        # Check if the difference is between 1 and 3
        if diff < 1 or diff > 3:
            return False
    
    # Check if the report is strictly increasing or strictly decreasing
    increasing = all(report[i] < report[i + 1] for i in range(len(report) - 1))
    decreasing = all(report[i] > report[i + 1] for i in range(len(report) - 1))
    
    # The report is safe if it is either strictly increasing or strictly decreasing
    return increasing or decreasing

def is_safe_with_one_removal(report):
    # Try removing each level one by one and check if it becomes safe
    for i in range(len(report)):
        modified_report = report[:i] + report[i+1:]
        if is_safe_report(modified_report):
            return True
    return False

# Read the input file
with open(r"C:\Users\Shobhnik\Downloads\AOC_DATA_DAY2.txt", 'r') as file:
    reports = []
    for line in file:
        # Convert each line to a list of integers
        report = list(map(int, line.split()))
        reports.append(report)

# Count the number of safe reports considering the Problem Dampener
safe_count = 0
for report in reports:
    if is_safe_report(report) or is_safe_with_one_removal(report):
        safe_count += 1

print(safe_count)
```
